## Task 1: Implement classification models on Chronic Kidney Disease dataset 

```python
# Load the libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler,LabelEncoder,OrdinalEncoder
from scipy.io import arff
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score,recall_score,mean_squared_error,precision_score,f1_score
from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.neighbors import KNeighborsClassifier
from sklearn.impute import SimpleImputer
```

```python
a="""Age(numerical) age in years
Blood Pressure(numerical) bp in mm/Hg
Specific Gravity(nominal) sg - (1.005,1.010,1.015,1.020,1.025)
Albumin(nominal) al - (0,1,2,3,4,5)
Sugar(nominal) su - (0,1,2,3,4,5)
Red Blood Cells(nominal) rbc - (normal,abnormal)
Pus Cell (nominal) pc - (normal,abnormal)
Pus Cell clumps(nominal) pcc - (present,notpresent)
Bacteria(nominal) ba - (present,notpresent)
Blood Glucose Random(numerical) bgr in mgs/dl
Blood Urea(numerical) bu in mgs/dl
Serum Creatinine(numerical) sc in mgs/dl
Sodium(numerical) sod in mEq/L
Potassium(numerical) pot in mEq/L
Hemoglobin(numerical) hemo in gms
Packed Cell Volume(numerical)
White Blood Cell Count(numerical) wc in cells/cumm
Red Blood Cell Count(numerical) rc in millions/cmm
Hypertension(nominal) htn - (yes,no)
Diabetes Mellitus(nominal) dm - (yes,no)
Coronary Artery Disease(nominal) cad - (yes,no)
Appetite(nominal) appet - (good,poor)
Pedal Edema(nominal) pe - (yes,no)
Anemia(nominal) ane - (yes,no)
Class (nominal) class - (ckd,notckd)
"""
l=[]
a=a.split(sep='\n')
import re
for i in a:
    l.append(re.findall("[a-zA-Z]*",i))
cols=[]
for i in l[:-1]:
    if(i[2]=='nominal' or i[2]=='numerical'):
        cols.append(i[0])
    else:
        cols.append(i[0]+" "+i[2]+" "+i[4]+" "+i[6])
```

```python
for i in range(len(cols)):
    cols[i]=cols[i].replace('nominal','')
    cols[i]=cols[i].replace('numerical','')
    cols[i]=cols[i].rstrip()
cols[-1]=cols[-1].split('  ')[0]
```

```python
cols
```

```python
# Load the dataset 
# Shuffel the dataset
df=pd.read_csv('chronic_kidney_disease.csv',header=None)
df.columns=cols
df
```

```python
df.replace('\t','',inplace=True,regex=True)
df.replace(' ','',inplace=True,regex=True)
for i in df.columns:
    df[i][df[i]=='?']=pd.to_numeric(df[i][df[i]=='?'], errors='coerce')
```

```python
for i in df.columns:
    print(df[i].value_counts(dropna=False))
    print()
```

```python
df.iloc[:,[i for i in range(0,5)]+[i for i in range(9,18)]]=imp.fit_transform(df.iloc[:,[i for i in range(0,5)]+[i for i in range(9,18)]])
df.columns=cols
```

```python
label=list(set(i for i in range(25)) - set([i for i in range(0,5)]+[i for i in range(9,18)]))
label
```

```python
df.replace(np.nan,'',inplace=True)
```

```python
# Preprocessing
# Encoding categorical variables (if any)
# Feature Scaling
# Filling missing values (if any)
enc=LabelEncoder()
for i in label:
    df.iloc[:,i]=enc.fit_transform(df.iloc[:,i])
imp=SimpleImputer(missing_values=0)
df.iloc[:,:-1]=imp.fit_transform(df.iloc[:,:-1])
df
```

```python
# Divide the dataset to training and testing set
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class',axis=1), df['Class'], test_size=0.30, random_state=42)
```

```python
# Implement various classification models such as SVM, Logistic regression, Decision tree and KNN, respectively.
# Note all models can be obtained from sklearn 
model=SVC()
model1=LogisticRegression()
model2=DecisionTreeClassifier()
model3=KNeighborsClassifier()
```

```python
# Train and test the models
model.fit(X_train,y_train)
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
```

## Task 2: Ensembling of classificatino models

```python
# Obtain the results of various models from Task 1
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
print()
print(model1.score(X_train,y_train))
print(model1.score(X_test,y_test))
print()
print(model2.score(X_train,y_train))
print(model2.score(X_test,y_test))
print()
print(model3.score(X_train,y_train))
print(model3.score(X_test,y_test))
print()
```

```python
from scipy.stats import mode
modelstrain=np.vstack((model.predict(X_train),model1.predict(X_train),model2.predict(X_train),model3.predict(X_train)))
majvottrain=np.array(mode(modelstrain)[0])
```

```python
# Apply ensembling of various classification models such as SVM, Logistic regression, Decision tree and KNN
modelstest=np.vstack((model.predict(X_test),model1.predict(X_test),model2.predict(X_test),model3.predict(X_test)))
majvottest=np.array(mode(modelstest)[0])
```

```python
print(accuracy_score(y_train,majvottrain.T))
print(accuracy_score(y_test,majvottest.T))
```

```python
# Build training analysis graphs for various parameters i.e., accuracy, precision, recall and f-measure for all models. 
import seaborn as sns
import matplotlib.pyplot as plt
models=[model,model1,model2,model3]
scores={'Models':['SVM','Logistic Regression','Decision Tree','KNN'],
        'accuracy':[accuracy_score(y_train,i.predict(X_train)) for i in models]
        ,'recall':[recall_score(y_train,i.predict(X_train)) for i in models]
        ,'precision':[precision_score(y_train,i.predict(X_train)) for i in models]
        ,'f1-score':[f1_score(y_train,i.predict(X_train)) for i in models]}
data=pd.DataFrame(scores)
df = pd.melt(data, id_vars="Models", var_name="scores", value_name="Scores")
plt.style.use('seaborn')
plot=sns.catplot(x="Models",y="Scores",hue="scores",kind='bar',data=df)
plot.set(ylim=(0.6))
plt.title('Training Accuracies')
```
