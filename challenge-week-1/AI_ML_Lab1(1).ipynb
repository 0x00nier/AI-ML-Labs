#**Task 1**

###Task 1.1

```python
import pandas as pd
import numpy as np
import matplotlib
df=pd.read_csv("/content/challenge-week-1/data/structured_data/data.csv")
df["Age"]=df["Age"].replace(np.NaN,df["Age"].mean())
df["Salary"]=df["Salary"].replace(np.NaN,df["Salary"].mean())
print(df)
#checking for git update
```

###Task 1.2

```python

df=df.astype({"Country":'category',"Purchased":'category',"Price Category Of Purchase":'category'})
print("Categorical Columns - ")
print(*df.select_dtypes(include='category').columns.tolist(),sep=', ')
print()
print("Numerical Columns - ")
print(*df.select_dtypes(exclude='category').columns.tolist(),sep=', ')
```

###Task 1.3

```python
print(df.head(10))
```

###Task 1.4

```python
a=pd.get_dummies(df["Country"],prefix="Country")
b=pd.get_dummies(df["Price Category Of Purchase"],prefix="Price Category Of Purchase")
df=pd.concat([df,a],axis=1)
df=pd.concat([df,b],axis=1)
df=df.drop(["Country","Price Category Of Purchase"],axis=1)
df['Purchased']=df['Purchased'].apply(lambda x: 1 if x=="Yes" else 0)
print(df.head(10))
```

###Task 1.5

```python
df[["Age","Salary"]]=(df[["Age","Salary"]]-df[["Age","Salary"]].mean())/df["Age"].std()
print(df.head())
```

###Task 3.6 (Final Task for Sanity Checking)

```python
import glob,os
from PIL import Image
c=0
viddim=[]
for infile in glob.glob("/content/challenge-week-1/data/videos/VID_*.mp4"):
  cap= cv2.VideoCapture(infile)
  i=0
  while(cap.isOpened()):
      ret, frame = cap.read()
      if ret == False:
          break
      cv2.imwrite('/content/'+'Vid'+str(c+1)+'/Vid'+str(c)+'_'+'Frame'+str(i)+'.jpg',frame)
      i+=1
  
  cap.release()
  cv2.destroyAllWindows()
  c+=1


```

```python
#w,h=Image.open("/content/Vid1/*.jpg").size
viddims=[]
import os
for i in range(1,15):
  directory = r'/content/Vid'+str(i)
  video=[]
  for filename in os.listdir(directory):
      if filename.endswith(".jpg") or filename.endswith(".png"):
          w,h=Image.open(os.path.join(directory, filename)).size
          video.append([w,h])
      else:
          continue
  viddims.append(video)    
```

```python
print(len(viddims[13]))
```

```python
def check(lst): 
    ele = lst[0] 
    chk = True
    for item in lst: 
        if ele != item: 
            chk = False
            break; 
              
    if (chk == True): print("Yes") 
    else: print("No")
c=1
for i in viddims:
  print("Vid"+str(c)+" has consistent dimensions? ")
  check(i)
  c+=1
```
